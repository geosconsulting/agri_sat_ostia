```python
list_files = glob.glob("../agri_sat_ostia/clipped_images/T32TQM_20210301/*.tif")
list_files
```

```python
list_files_add = []
for f in list_files:
    fs = f.split('\\')[-1].split('.')[0]    
    result = re.search('_10m$',fs)
    if result:
        list_files_add.append(f)
list_files_add        
```

```python
# time_names= [file_time.split('\\')[1][7:15] for file_time in list_files_add]
# time_names = datetime.strptime(time_names[0],'%Y%m%d')
```

```python
with gw.config.update(sensor='s210'):
    with gw.open(list_files_add, stack_dim='band') as src_stack: #,time_names=time_names
        print(src_stack.band)
        print(src_stack.gw.bounds)
        print(src_stack.crs)
        for k, v in src_stack.gw.config.items():
            print('Keyword:', k.ljust(15), 'Value:', v)
```

```python
src_stack
```

```python
fig, ax = plt.subplots(dpi=200)
src_stack.where(src_stack != 0).sel(band=['red','green', 'blue']).gw.imshow(robust=True, ax=ax)
plt.tight_layout(pad=1)
```
